### I / A : Import libraries

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

### I / B : Importing our data

```python
inference_df = pd.read_csv('data/eda_clean_df.csv')
inference_df
```

### I / C : Linear Regression using statsmodel

```python
inference_drop = inference_df.loc[:, inference_df.columns != 'ocean_proximity']
# inference_drop
```

```python
X = inference_drop.loc[:, inference_drop.columns != 'median_house_value']
y = inference_df.loc[:, inference_df.columns == 'median_house_value']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X)
```

```python
results = model.fit()
results.params
```

```python
print(results.summary())
```
